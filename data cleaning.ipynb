```python
deflator_file = 'Indiadeflator.csv'
deflator_data = pd.read_csv(deflator_file)
```

```python
df2 = pd.DataFrame(deflator_data)
df2 = df2.set_index('Year')
df2
```

```python
deflator = []
for i in year:
    if i == 2018:
        deflator.append(None)
    else:
        deflator.append(float(df2.loc[i]))
print(deflator)
```

```python
df1['Deflator'] = deflator
cols = list(df1.columns)
print(cols)
```

```python
df1 = df1[['sa_finance1_cocode', 'sa_company_name', 'sa_finance1_year', 'year', 'Deflator', 'sa_ann_rep_months', 'sa_fs_format', 'sa_sales', 'sa_interest_exp', 'sa_int_exp_lt_funds', 'sa_int_exp_st_fund', 'sa_total_interest_exp', 'sa_int_exp_st_pc_avg_st_borr', 'sa_int_exp_lt_pc_avg_lt_borr', 'sa_long_term_borrowings', 'sa_short_term_borrowings', 'sa_borrowings', 'sa_curr_portion_lt_borr', 'sa_total_assets']]
df1
```

```python
df_ratio = df1[['sa_finance1_cocode', 'sa_company_name', 'year', 'Deflator', 'sa_int_exp_lt_funds', 'sa_int_exp_st_fund']]
df_ratio.head() #dataframe for ST/LT
```

```python
df_ratio1 = df_ratio[pd.notnull(df_ratio.sa_int_exp_st_fund)] #dataframe dropping null ST values
df_ratio2 = df_ratio1[pd.notnull(df_ratio.sa_int_exp_lt_funds)]
df_ratio2 #dataframe without any null ST, LT values
```

```python
ratio = df_ratio2['sa_int_exp_st_fund']/df_ratio2['sa_int_exp_lt_funds']
df_ratio2['ratio'] = ratio
df_ratio2 #adding ratio column
```

```python
mean = []
for i in df_ratio2.sa_finance1_cocode:
    #name = df_ratio2.sa_company_name
    index = df_ratio2.sa_finance1_cocode[df_ratio2.sa_finance1_cocode == i].index
    mean_st_lt = np.mean(df_ratio2.loc[index].ratio)
    mean.append(mean_st_lt)
df_ratio2['mean'] = mean #adding mean of ST/LT ratio column
df_ratio2.head()
```

```python
df_ratio2_nd = df_ratio2.drop_duplicates(subset='sa_company_name', keep='first') #dropping duplicates
df_ratio2_nd.head()
```

```python
df_ratio_plot = df_ratio2_nd.sort_values(by=['mean']) #order df in ascending values of mean
df_ratio_plot = df_ratio_plot.set_index('sa_company_name') #setting company as index
df_ratio_plot = df_ratio_plot[['mean']]
df_ratio_plot #company and ST/LT mean over all available years - need to add year filter in earlier df
```
